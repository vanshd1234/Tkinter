```python
%%writefile bank.py
import tkinter as tk
import json
import os


def sign_up():
    a='login id'
    b='password'
    c='acc'
    text1=ent_var1.get()
    text2=ent_var2.get()
    acc=ent_var3.get()
    #print(a + text1)
    #print(b + text2)
    ent_var1.set('')
    ent_var2.set('')
    ent_var3.set('')
    d={c:acc,a:text1,b:text2}
    print(d)
    f=open(f'bank_info/{acc}.json','w')
    json.dump(d,f)
    f.close()

def login_up():
    acc=ent_var3.get()
    acc=acc+'.json'
    if acc in os.listdir('bank_info'):
        f=open(f"bank_info/{acc}")
        data=json.load(f)
        if ent_var2.get()==data['password']:
            print('login sucessfull')
            #login
            win=tk.Toplevel(root)
            win.focus()
            win.grab_set()
            lab=tk.Label(win,text='login sucessfull')
            lab.pack()
            
            lab1=tk.Label(win,text=(f"ACCOUNT NO. : {data['acc']}"))
            lab1.pack()
            
            lab1=tk.Label(win,text=(f"LOGIN ID : {data['login id']}"))
            lab1.pack()

            #buttom button for login
            b=tk.Button(win,text='exit',command=win.destroy)
            b.pack(side=tk.BOTTOM)
            #title
            win.title('login info')
            win.wm_minsize(400,400)
            win.wm_maxsize(450,450)
            
            #debit
            b1=tk.Button(win,text='DEBIT',)
            b1.pack()
            
            #credit
            b2=tk.Button(win,text='CREDIT',)
            b2.pack()



        
        else:
            win=tk.Toplevel(root)
            win.focus()
            win.grab_set()
            lab2=tk.Label(win,text='Login Failed')
            lab2.pack()


    
root = tk.Tk()

ent_var1=tk.StringVar()
ent_var2=tk.StringVar()
ent_var3=tk.StringVar()

frame =tk.Frame(root, bg='YELLOW')
clb=tk.Label(root, text='BANK INFO')
clb.pack(side=tk.TOP , fill = tk.X)

#account info
l1=tk.Label(root,text='ACCOUNT : ',)
l1.pack(side=tk.TOP)

ent_box3 = tk.Entry(root,font=('Times',14, ),textvariable=ent_var3)
ent_box3.pack()
ent_box3.focus()

#login info
l=tk.Label(root,text='LOGIN ID :',)
l.pack(side=tk.TOP)

ent_box1 = tk.Entry(root,font=('Times',14, ),textvariable=ent_var1)
ent_box1.pack()

#password info
l1=tk.Label(root,text='PASSWORD : ')
l1.pack(side=tk.TOP)

ent_box2 = tk.Entry(root,font=('Times',14),textvariable=ent_var2)
ent_box2.pack()

ent_box1.bind('<Return>',sign_up)
ent_box2.bind('<Return>',sign_up)
ent_box3.bind('<Return>',sign_up)

#login button
ent_button1 = tk.Button(root, text='login', command=login_up)
ent_button1.pack()

ent_button2 = tk.Button(root, text='signup',command=sign_up)
ent_button2.pack()

b=tk.Button(root,text='exit',command=root.destroy)
b.pack()


root.wm_minsize(400,400)

root.title('Bank Application')


root.mainloop()
```
